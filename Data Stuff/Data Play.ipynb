```python
import pandas as pd
import numpy as np
```

```python
#Read all files as pandas dataframes
summary = pd.read_csv('summary.csv')
host = pd.read_csv('host.csv')
rvbsm = pd.read_csv('rvbsm.csv')
spat = pd.read_csv('spat.csv')

#Drop collumns that are all null
summary = summary.dropna(axis=1, how='all')
host = host.dropna(axis=1, how='all')
rvsbm = rvbsm.dropna(axis=1, how='all')
spat = spat.dropna(axis=1, how='all')

#Drop key column from all dataframes
summary = summary.drop(['key','uuid'], axis=1)
host = host.drop(['key','uuid'], axis=1)
rvbsm = rvbsm.drop(['key'], axis=1)
spat = spat.drop(['key','uuid'], axis=1)


#Get 25%  samples of each dataframe
# summary_sample = summary.sample(frac=0.25)
# host_sample = host.sample(frac=0.25)
# rvbsm_sample = rvbsm.sample(frac=0.25)
# spat_sample = spat.sample(frac=0.25)


# summary_sample.set_index('device', inplace=True)
summary[summary['numwarnings'] >0]
```

```python
merge_trip = pd.merge(rvbsm, spat, how = 'left', on = ['device','filedate','filetimeutc', 'localtimems'])
merge_trip = pd.merge(merge_trip, host, how = 'left', on = ['device','filedate','filetimeutc', 'localtimems'])
merge_trip.drop(['intersectid','laneid', 'minendtime', 'movementphase', 'signalgroupid'], axis=1, inplace=True)
merge_trip = merge_trip.dropna()
# merge_trip.info()

trips =[]
for group in merge_trip.groupby(['device','filedate','filetimeutc', 'localtimems']):
    trips.append(group[1])

data = trips[0]
for i in range(len(trips)):
    data = pd.concat([data, trips[i]])

data.drop(['filedate','filetimeutc', 'localtimems', 'device','basicvehicleclass'], axis=1, inplace=True)
data['rear_end'] = np.random.randint(0, 10, data.shape[0])>7 

data.to_csv('data.csv', index=False)

data
```

```python
#split data into train and test

def dist(x1,y1,x2,y2):
    return np.sqrt((x1-x2)**2 + (y1-y2)**2)

from sklearn.model_selection import train_test_split
data['rear_end'] = np.random.randint(0, 10, data.shape[0])>7 
y = data.pop('rear_end')
X = data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

# train RandomForestClassifier model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(max_depth=15, random_state=0)
model.fit(X_train, y_train)

from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(model.score(X_test, y_test))

#export model using joblib
import joblib
joblib.dump(model, 'model.pkl')

```
